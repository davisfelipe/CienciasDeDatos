# EJERCICIO 1
Crear un ejemplo de pipe con tres funciones.  La primera y la tercera función deben tener argumentos diferentes al dataframe

```python
def vectorial(dataframe,cola,colb):
    dataframe['A']=np.sqrt(dataframe[cola]**2+dataframe[colb]**2)
    return dataframe
def signo(dataframe):
    return dataframe*-1
def exponencial(dataframe,cola):
    dataframe[cola]=np.exp(dataframe[cola])
    return dataframe
display(dataNumeros.pipe(vectorial,'B','C').pipe(signo).pipe(exponencial,'E'))
```

# EJERCICIO 2
