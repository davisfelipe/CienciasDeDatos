```python
#Leer el archivo y guardarlo como archivo de pandas
icfes=pd.read_csv('saber.csv')
```

## Limpieza de datos

### df.isnull()
Toma un dataframe y retorna un dataframe donde los datos corresponden a valores Booleanos donde True representan un valor nulo y False no nulo.

```python
display(icfes['ESTU_COLE_TERMINO'].isnull())
```

### df.dropna()
La funcion elimina los registros del dataframe donde se encuentren valores nullos en alguna de sus columnas, dicho comportamiento se modifica dependiendo de los parametros que se le pasen a la funcion:
* **axis:** Cuando se envia el parametro 1 elimina las columnas donde se presente algun nulo, en el caso de 0, elimina el registro/fila donde se presente el nulo.
* **how:** Elimina el registro dependiendo del tipo de parametro, el en caso que se envie un *"all"* elimina el registro siempre y cuando todos sus campos posean un null, si se envia un *"any"* elimina el registro si encuentra almenos 1 null

```python
display(icfes.dropna(axis=1))
```

### df.fillna()
La funcion permite rellenar los valores nullos encontrados en el dataframe respecto a todo el dataframe o simplemente con respecto a una columna, esto dependera del tipo de dato que se encuentre en cada una de las columnas

```python
icfes['ESTU_TITULOOBTENIDOBACHILLER'].fillna(value='Bachiller')
```

```python
icfes['PUNT_GLOBAL'].fillna(0)
display(icfes['MOD_RAZONA_CUANTITAT_PUNT'].fillna(value=icfes['MOD_RAZONA_CUANTITAT_PUNT'].mean()))
```

## Seleccion de datos

```python
#Returna True o False para aquellos puntajes que se encuentren por encima de la media del puntaje global
icfes['PUNT_GLOBAL']>=icfes['PUNT_GLOBAL'].mean()
```

```python
#Retorna todos los datos correspondientes a la filtracion
icfes[icfes['PUNT_GLOBAL']>=(icfes['PUNT_GLOBAL'].mean())]
```

```python
#Retorna datos especificos (documento, genero, municipio) del filtro realizado respecto a los puntajes globales mayores
#a la media
icfes[['ESTU_TIPODOCUMENTO','ESTU_GENERO','ESTU_MCPIO_RESIDE']][icfes['PUNT_GLOBAL']>=(icfes['PUNT_GLOBAL'].mean())]
```

```python
#Retorna datos especificos (tipo de documento, genero, municipio, puntaje) del filtro realizado respecto a los estudiantes
#que poseen un puntaje mayor al promedio y que sean mujeres
icfes[['ESTU_TIPODOCUMENTO','ESTU_GENERO','ESTU_MCPIO_RESIDE','PUNT_GLOBAL']][(icfes['PUNT_GLOBAL']>=icfes['PUNT_GLOBAL'].mean()) & (icfes['ESTU_GENERO']=='F')]
```

## Consulta por metodos

### where ()
Retorna los valores que cumplan con la condicion de la funcion, de tal forma que los registros que no cumplan con esta, son remplazados con valores NaN por defecto, dicho parametro se puede modificar con el parametro *"other"*


```python

icfes['PUNT_GLOBAL'].where(icfes['PUNT_GLOBAL']>=icfes['PUNT_GLOBAL'].mean(),other=0)
```

### mask ()
Retorna los valores que no cumplen con la condicion de la funcion, de tal forma que los registros que cumplan con esta, son remplazados con valores NaN por defecto, dicho parametro se puede modificar con el parametro *"other"*

```python
icfes['PUNT_GLOBAL'].mask(icfes['PUNT_GLOBAL']>=icfes['PUNT_GLOBAL'].mean(),other='Sobre la media')
```

### isin ()
Retorna un dataframe de booleanos en donde dado un conjunto de vectores retornara aquellos que se encuentren dentro de estos parametros

```python
icfes[['ESTU_TIPODOCUMENTO','ESTU_GENERO','ESTU_MCPIO_RESIDE']][icfes['ESTU_MCPIO_RESIDE'].isin(['NEIVA','BARRANQUILLA','CALI'])]
```

### query ()
la funcion permite ingresar directamente condiciones donde retornara los registros que cumplan dichas condiciones

```python
icfes.query("ESTU_GENERO == 'M'").query("ESTU_TIPODOCUMENTO!='CC'")
```
